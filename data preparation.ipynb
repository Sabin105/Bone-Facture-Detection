```python
import pandas as pd 
```

```python
import os 
```

```python
fact_img=r'Image/Augmented Dataset'
```

```python
# complete path 
facture_image_path=[]
```

```python
lables = os.listdir(fact_img)
lables
```

```python
# get dictionary with label names 
label_path_dict=dict.fromkeys(lables)
label_path_dict
```

```python
for label in label_path_dict.keys():
    lbl_path=os.path.join(fact_img,label)
    img_path=[os.path.join(lbl_path,img) for img in os.listdir(lbl_path)]
    label_path_dict[label]=img_path 
  
```

```python
label_path_dict['Fractured']=label_path_dict['Fractured'][1:]
```
